```python
files = [
    'chantiers-a-paris-copie2019.csv',
    'chantiers-a-paris-copie2020.csv',
    'chantiers-a-paris-copie2021.csv',
    'chantiers-a-paris-copie2022.csv'
]
# Charger tous les fichiers dans une liste de DataFrames
dataframes = [pd.read_csv(file, sep=';') for file in files]

# Concaténer tous les DataFrames en un seul
combined_df = pd.concat(dataframes, ignore_index=True)
# Filtrer les lignes
filtered_df = combined_df[combined_df["Maîtrise d'ouvrage principale"].str.contains('SNCF|RATP', na=False)]
# Sauvegarder le résultat dans un nouveau fichier CSV
filtered_df.to_csv('TravauxSNCF_RATP_2019-2022.csv', index=False, sep=';')

print("Le fichier 'TravauxSNCF_RATP_2019-2022.csv' a été créé avec succès.")
```
