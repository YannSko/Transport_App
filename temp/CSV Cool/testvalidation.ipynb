```python
# Lire le fichier CSV
df = pd.read_csv('chantiers-a-paris-copie2019.csv', sep=';')

# Compter le nombre d'occurrences de 'SNCF' dans la colonne 'Maîtrise d'ouvrage principale'
sncf_count = df["Maîtrise d'ouvrage principale"].str.contains('SNCF', na=False).sum()

# Afficher le résultat
print(f"Nombre d'occurrences de 'SNCF' : {sncf_count}")

sncf_count = df["Maîtrise d'ouvrage principale"].str.contains('RATP', na=False).sum()
print(f"Nombre d'occurrences de 'RATP' : {sncf_count}")
```
